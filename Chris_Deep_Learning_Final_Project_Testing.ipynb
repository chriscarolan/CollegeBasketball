```python
testing_combined = testing_combined.rename(columns={"Year_x": "Year"})
testing_combined = testing_combined.rename(columns={"Round_x": "Round"})
testing_combined = testing_combined.rename(columns={"Region Number_x": "Region Number"})
testing_combined = testing_combined.rename(columns={"Region Name_x": "Region Name"})
```

```python
testing_combined = testing_combined.drop([#'Team_x', 'Winning Team', 'Losing Team', 'Team_y','Team_B_x', 'Team_B_y',
                                          'Winning Score_x', 'Losing Score_x',
                                          'Winning Seed_x', 'Losing Seed_x', 'Season_x', 'seed_x',
                                          'Year_y', 'Round_y', 'Region Number_y', 'Seed_A_y', 'Score_A_y',
                                          'Score_B_y', 'Seed_B_y', 'Winning Seed_y', 'Losing Seed_y',
                                          'Season_y', 'seed_y', 'Region Name_y', 'Winning Team_y',
                                          'Winning Team_y', 'Winning Score_y', 'Losing Score_y', 'Losing Team_y',
                                          'Region Name', 'Region Number'], axis=1)
```

```python
# seed and score
testing_combined = testing_combined.rename(columns={"Seed_A_x": "Team A Seed"})
testing_combined = testing_combined.rename(columns={"Score_A_x": "Team A Score"})
testing_combined = testing_combined.rename(columns={"Seed_B_x": "Team B Seed"})
testing_combined = testing_combined.rename(columns={"Score_B_x": "Team B Score"})

# winning/losing team
testing_combined = testing_combined.rename(columns={"Winning Team_x": "Winning Team"})
testing_combined = testing_combined.rename(columns={"Losing Team_x": "Losing Team"})

# team A tempo
testing_combined = testing_combined.rename(columns={"Tempo_x": "Team A Tempo"})
testing_combined = testing_combined.rename(columns={"RankTempo_x": "Team A Tempo Rank"})
testing_combined = testing_combined.rename(columns={"AdjTempo_x": "Team A Adjusted Tempo"})
testing_combined = testing_combined.rename(columns={"RankAdjTempo_x": "Team A Adjusted Tempo Rank"})

# team A offense
testing_combined = testing_combined.rename(columns={"OE_x": "Team A OE"})
testing_combined = testing_combined.rename(columns={"RankOE_x": "Team A OE Rank"})
testing_combined = testing_combined.rename(columns={"AdjOE_x": "Team A Adjusted OE"})
testing_combined = testing_combined.rename(columns={"RankAdjOE_x": "Team A Adjusted OE Rank"})

# team A Defense
testing_combined = testing_combined.rename(columns={"DE_x": "Team A DE"})
testing_combined = testing_combined.rename(columns={"RankDE_x": "Team A DE Rank"})
testing_combined = testing_combined.rename(columns={"AdjDE_x": "Team A Adjusted DE"})
testing_combined = testing_combined.rename(columns={"RankAdjDE_x": "Team A Adjusted DE Rank"})

# team A Efficiency Margins
testing_combined = testing_combined.rename(columns={"AdjEM_x": "Team A EM"})
testing_combined = testing_combined.rename(columns={"RankAdjEM_x": "Team A EM Rank"})
```

```python
n_splits = 5
kfold = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
```

```python
input_shape = X_train_reshaped.shape[1:]
```

```python
y_train.reset_index(drop=True, inplace=True)
fold = 0
results = []
for train_index, val_index in kfold.split(X_train_reshaped, y_train):
    print(f'Fold {fold + 1}/{n_splits}')

    X_train_fold, X_val_fold = X_train_reshaped[train_index], X_train_reshaped[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    model = create_model(input_shape)

    callbacks = [
    ModelCheckpoint(f"best_model_fold_{fold}.keras", save_best_only=True, monitor="val_loss"),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=20, min_lr=0.1),
    EarlyStopping(monitor="val_loss", patience=50, verbose=1)
]

    history = model.fit(
        X_train_fold, y_train_fold,
        validation_data=(X_val_fold, y_val_fold),
        batch_size=batch_size,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )

    loss, accuracy = model.evaluate(X_val_fold, y_val_fold, verbose=0)
    print(f'Validation Accuracy: {accuracy}')

    results.append(accuracy)
    fold += 1
```
